# Overall Hallucination

```python
df['feedback_scores.hallucination_metric'] = pd.to_numeric(df['feedback_scores.hallucination_metric'], errors='coerce')
df_ft['feedback_scores.hallucination_metric'] = pd.to_numeric(df_ft['feedback_scores.hallucination_metric'], errors='coerce')

average_hallucination_score = df['feedback_scores.hallucination_metric'].mean()
average_hallucination_score_ft = df_ft['feedback_scores.hallucination_metric'].mean()

print(f"Rata-rata Skor Halusinasi Base Model: {average_hallucination_score:.2f}")
print(f"Rata-rata Skor Halusinasi Fine-Tuned Model: {average_hallucination_score_ft:.2f}")

change = average_hallucination_score_ft - average_hallucination_score
percentage_change = (change / average_hallucination_score) * 100

if change < 0:
    print(f"Persentase penurunan skor halusinasi: {abs(percentage_change):.2f}%")
else:
    print(f"Persentase kenaikan skor halusinasi: {percentage_change:.2f}%")
```

## Pola Kegagalan Utama: Fabrikasi dan "Eskalasi Medis"

Ini adalah pola yang paling dominan, terjadi di hampir semua kasus. Model menerima konteks yang berisi beberapa kemungkinan penyebab atau saran umum, namun alih-alih menggunakannya, model justru membuat daftar penyakit baru yang lebih panjang, lebih spesifik, dan seringkali lebih serius.

### Kasus #1 (Sakit Perut Kanan)
- **Kesalahan**: Konteks mengaitkan gejala dengan maag dan konstipasi, tetapi chatbot mengabaikannya dan memunculkan diagnosis baru seperti kolonitis dan penyakit radang usus (IBD).

### Kasus #3 (Sakit Kepala & Demam)
- **Kesalahan**: Konteks sangat berhati-hati dan hanya menyarankan konsultasi dokter. Chatbot melakukan eskalasi dengan menyebutkan daftar panjang penyakit spesifik seperti gastritis, demam tifoid, demam berdarah, dll.

### Kasus #4 (Nyeri Dada Kiri)
- **Kesalahan**: Pola yang sama terulang. Konteks hanya memberikan saran umum, tetapi chatbot memunculkan kemungkinan penyebab serius seperti serangan jantung, angina, pleuritis, dan pneumonia.

### Kasus #6, #7, #8, #10 (Berbagai Nyeri Perut)
- **Kesalahan**: Keempat kasus ini menunjukkan pola identik. Model diberikan gejala dan konteks yang terbatas, namun jawabannya selalu berupa fabrikasi daftar panjang kondisi medis (pankreatitis, batu ginjal, kanker lambung, penyakit celiac, hepatitis) yang tidak ada di dalam sumber.

**Inti Masalah**:  
Model ini bertindak seperti mesin pencari yang menampilkan *"top related searches"* untuk sebuah gejala, bukan sebagai asisten yang membaca dan menyarikan dokumen yang diberikan.

## Pola Kegagalan Sekunder: Kontradiksi dan Informasi Salah

Meskipun tidak sesering fabrikasi, model ini juga menunjukkan ketidakmampuan untuk memproses informasi yang benar atau memahami batasan.

### Kasus #5 (Karbohidrat untuk Maag)
- **Kesalahan**: Terjadi kontradiksi langsung. Konteks menyatakan *"belum ada bukti kuat"* untuk menghindari nasi atau jagung, tetapi chatbot justru menyarankan untuk membatasi konsumsi makanan tersebut.

### Kasus #2 (Tumor Pilorus)
- **Kesalahan**: Menunjukkan informasi salah dengan mendefinisikan tumor sebagai *"benjolan yang tidak bersifat ganas (kanker)"*, padahal tumor bisa bersifat ganas (malignant) atau jinak (benign). Selain itu, chatbot sepenuhnya mengabaikan pertanyaan inti pengguna tentang *"perut sering berbunyi."*

### Kasus #9 (Feses Biru)
- **Kesalahan**: Meskipun konteks sudah memberikan jawaban yang benar (pewarna makanan), chatbot tetap menambahkan penyebab lain yang tidak ada di sumber, seperti gangguan pencernaan dan efek obat-obatan.

## Kesimpulan Diagnosis untuk Model Base

Perilaku Model Base ini menunjukkan karakteristik khas dari LLM pra-terlatih yang belum di-fine-tune untuk tugas spesifik yang membutuhkan kepatuhan ketat pada konteks.

### Masalah Utama:
- **Kecenderungan untuk "Membantu Secara Berlebihan"**  
  Model berusaha memberikan jawaban selengkap mungkin dengan mengakses seluruh pengetahuannya, tanpa menyadari bahwa batasan konteks adalah aturan utama.

- **Kurangnya Kemampuan "Grounding"**  
  Model tidak memiliki mekanisme kuat untuk memprioritaskan informasi dari konteks di atas informasi dari data pra-pelatihannya.

- **Kesulitan Memahami Negasi dan Nuansa**  
  Dalam kasus kontradiksi (seperti kasus #5), model gagal memahami nuansa kalimat *"belum ada bukti kuat"* dan salah menafsirkannya sebagai larangan.


##  Pola Kegagalan Sekunder: Kontradiksi dan Misinformasi pada Model Finetuning

###  Pelanggaran Instruksi Negatif

- **Kasus #2**: Pengguna bertanya soal alergi dan maag.
  - **Konteks**: Melarang penyebutan nama obat.
  - **Kesalahan**: Model tetap menyebutkan nama obat seperti ranitidin, omeprazol.
  - **Jenis Halusinasi**: *Instruksi Dilanggar* → swamedikasi bisa terjadi.

##  Kesimpulan Umum

###  Diagnosis Masalah Model:

1. **Cenderung "Membantu Berlebihan"**  
   Model mencoba menyebut semua kemungkinan penyakit, tanpa memperhatikan konteks atau kebutuhan pengguna.

2. **Kurangnya Grounding pada Konteks**  
   Tidak bisa membedakan mana informasi dari konteks yang harus diprioritaskan dibanding pengetahuan umum dari pre-training.

3. **Tidak Tahan terhadap Instruksi Negatif**  
   Model tidak bisa berkata "tidak tahu" atau "tidak boleh menjawab itu", walaupun sudah ada larangan eksplisit.
