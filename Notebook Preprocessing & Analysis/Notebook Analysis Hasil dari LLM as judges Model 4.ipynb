# Overall Hallucination

```python
df['feedback_scores.hallucination_metric'] = pd.to_numeric(df['feedback_scores.hallucination_metric'], errors='coerce')
df_ft['feedback_scores.hallucination_metric'] = pd.to_numeric(df_ft['feedback_scores.hallucination_metric'], errors='coerce')

average_hallucination_score = df['feedback_scores.hallucination_metric'].mean()
average_hallucination_score_ft = df_ft['feedback_scores.hallucination_metric'].mean()

print(f"Rata-rata Skor Halusinasi Base Model: {average_hallucination_score:.2f}")
print(f"Rata-rata Skor Halusinasi Fine-Tuned Model: {average_hallucination_score_ft:.2f}")

change = average_hallucination_score_ft - average_hallucination_score
percentage_change = (change / average_hallucination_score) * 100

if change < 0:
    print(f"Persentase penurunan skor halusinasi: {abs(percentage_change):.2f}%")
else:
    print(f"Persentase kenaikan skor halusinasi: {percentage_change:.2f}%")
```

```python
print('Treshold dengan Hallucination dengan hallcuintion diatas 0.5:')

bad_answers_count = df[df['feedback_scores.hallucination_metric'] > 0.5].shape[0]
print(f"Jumlah jawaban dengan  halusinasi tinggi: {bad_answers_count} dari 100")

# Hitung jumlah jawaban dengan skor di bawah 0.5
bad_answers_count_ft = df_ft[df_ft['feedback_scores.hallucination_metric'] > 0.5].shape[0]
print(f"Jumlah jawaban dengan  halusinasi tinggi (fine tuned): {bad_answers_count_ft} dari 100")
```

### 1. Kontradiksi Langsung terhadap Konteks (Kesalahan Paling Fatal)
Ini adalah jenis halusinasi yang *paling jelas dan fatal*, di mana model menyatakan sesuatu yang secara eksplisit dibantah oleh teks konteks yang diberikan.

* *Kasus #2 (Nyeri Pasca Usus Buntu):* Model membuat klaim yang bertentangan langsung dengan peringatan dalam konteks.
    > *Klaim Model:* Menyatakan bahwa usus buntu bisa kambuh atau keluhan saat ini berkaitan dengan operasi sebelumnya.
    > *Fakta Konteks:* Secara eksplisit menyatakan, "keluhan yang anda alami saat ini bukan berkaitan dengan apendisitis" dan kemungkinan besar "tidak berhubungan dengan riwayat operasi."

* *Kasus #5 (Ambeien):* Model merekomendasikan pengobatan yang secara tegas dilarang oleh konteks.
    > *Klaim Model:* Merekomendasikan bawang putih sebagai pengobatan ambeien.
    > *Fakta Konteks:* Secara tegas menyatakan, "penggunaan bawang putih tidak mengobati ambeien yang anda derita, jadi berita ini tidak benar ya."

* *Kasus #7 (Yogurt):* Model menyangkal hubungan sebab-akibat yang dijelaskan dalam konteks.
    > *Klaim Model:* Mengklaim bahwa yogurt tidak menyebabkan asam lambung.
    > *Fakta Konteks:* Menyebutkan bahwa konsumsi yogurt berlebihan justru bisa memicu asam lambung.

* *Kasus #9 (Nyeri Ulu Hati):* Sama seperti Kasus #2, model mengabaikan negasi yang ada di dalam konteks.
    > *Klaim Model:* Mengaitkan nyeri ulu hati dengan riwayat operasi pasien.
    > *Fakta Konteks:* Menyatakan bahwa kemungkinan besar keluhan "tidak berhubungan dengan riwayat operasi."

---

### 2. Fabrikasi Informasi Medis (Pola Paling Umum)
Ini adalah *pola halusinasi yang paling sering terjadi*. Model "mengarang" atau menambahkan informasi medis baru—seperti diagnosis, nama obat, atau tes medis—yang sama sekali tidak ada dalam teks konteks.

* *Kasus #1 (Demam & Maag):* Model memunculkan diagnosis baru yang tidak pernah disebutkan.
    > *Informasi Konteks:* Hanya membahas demam dan maag.
    > *Fabrikasi Model:* Menambahkan diagnosis *GERD* dan *dispepsia fungsional*.

* *Kasus #4 (Alergi & Maag):* Model melanggar aturan dengan menyebutkan nama obat spesifik.
    > *Informasi Konteks:* Secara umum membahas penanganan alergi dan maag, serta melarang keras penyebutan nama obat.
    > *Fabrikasi Model:* Menyebutkan nama obat seperti *Benadryl, Claritin, dan Ibuprofen*.

* *Kasus #6 (Sangobion & Diare):* Model mengarang efek samping atau bahaya baru yang tidak berdasar.
    > *Informasi Konteks:* Membahas hubungan Sangobion dengan warna feses dan diare.
    > *Fabrikasi Model:* Menambahkan klaim berbahaya bahwa Sangobion bisa "memicu perdarahan lebih lanjut."

* *Kasus #8 & #10 (Gejala Umum):* Model menambahkan daftar panjang kemungkinan penyakit yang tidak relevan dengan konteks.
    > *Informasi Konteks:* Membahas beberapa gejala umum.
    > *Fabrikasi Model:* Memunculkan daftar penyakit baru seperti *GERD, konstipasi, flu, sinusitis,* dll.

---

### 3. Memberikan Informasi yang Salah Secara Faktual
Jenis ini lebih berbahaya dari sekadar fabrikasi. Model tidak hanya mengarang informasi, tetapi informasi yang dikarangnya *salah secara fundamental atau ilmiah* dan berpotensi menyesatkan pengguna.

* *Kasus #3 (Feses Biru):* Ini adalah contoh paling representatif untuk kategori ini, di mana model memberikan penjelasan medis yang sepenuhnya keliru.
    > *Klaim Model (Salah):* Feses biru disebabkan oleh *bilirubin* dan *gangguan hati*.
    > *Fakta Konteks (Benar):* Feses biru disebabkan oleh *zat pewarna makanan*.

---

### 4. Memberikan Gambaran yang Tidak Lengkap dan Menyesatkan
Model gagal menyajikan informasi penting dari konteks. Meskipun jawabannya tidak sepenuhnya salah, *informasi yang hilang (omission) membuat jawaban akhir menjadi tidak lengkap dan menyesatkan*.

* *Kasus #2 (Nyeri Pasca Usus Buntu):* Selain melakukan kontradiksi, model juga melakukan halusinasi jenis ini.
    > *Kesalahan Model:* Gagal menyebutkan banyak kemungkinan penyebab nyeri lain (seperti *dispepsia, kolitis, hernia, dll*) yang sudah tercantum dengan jelas di dalam konteks, sehingga menyederhanakan masalah secara berlebihan.
